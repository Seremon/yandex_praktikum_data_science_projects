## Расчёты и добавление результатов в таблицу

*Посчитаем и добавим в таблицу цену квадратного метра:*

```python
data['price_per_square_meter'] = (data['last_price'] / data['total_area']).astype('int')
```

*Посчитаем и добавим в таблицу день недели, месяц и год публикации объявления:*

```python
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
```

```python
data['month_exposition'] = data['first_day_exposition'].dt.month
```

```python
data['year_exposition'] = data['first_day_exposition'].dt.year
```

*Посчитаем и добавим в таблицу этаж квартиры — первый, последний, другой: (используем функцию)*

```python
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    elif 1 < floor < floors_total:
        return 'другой'
    
data['floor_category'] = data.apply(floor_category, axis = 1)
```

*Посчитаем и добавим в таблицу соотношение жилой и общей площади, а также отношение площади кухни к общей:*

```python
data['living_area_to_the_total'] = data['living_area'] / data['total_area']
```

```python
data['kitchen_area_to_the_total'] = data['kitchen_area'] / data['total_area']
```
